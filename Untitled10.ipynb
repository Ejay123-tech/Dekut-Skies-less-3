```python
# Importing some necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Dropping some irelevant columns in the data
loan.drop(["ID","loan_limit","approv_in_adv","Credit_Worthiness","open_credit","rate_of_interest","Interest_rate_spread","Upfront_charges",
           "term","Neg_ammortization","interest_only","property_value","construction_type","occupancy_type","Secured_by","total_units","dtir1",
          "submission_of_application","LTV","Region","Security_Type","income" ],axis=1,inplace=True)
```

```python
# Checking for missing values in the data
loan.isnull().sum()
```

```python
# Handling missing value
loan["loan_purpose"]=loan["loan_purpose"].fillna(loan["loan_purpose"].mode())
loan["age"]=loan["age"].fillna(loan["age"].mode())
```

```python
# Drop missing values
loan.dropna(inplace=True)
```

### Hypothesis testing


```python
# Create a contingency table
contingency_table = pd.crosstab(loan['loan_purpose'], loan['Status'])
```

```python
# Perform the chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
```
